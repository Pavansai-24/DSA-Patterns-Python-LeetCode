```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://companiesmarketcap.com/"
response=requests.get(url)
Soup=BeautifulSoup(response.content,"html.parser")
```

```python
print(Soup.title.string)
```

```python
Soup.find("table")
```

```python
Soup.find_all("table")
```

```python
Soup.find("table",
          class_="default-table table marketcap-table dataTable")
```

```python
Soup.find_all("th")
```

```python
word_titles=Soup.find_all("th")
word_titles
```

```python
word_titles=[title.text.strip() for title in word_titles]
word_titles
```

```python
import pandas as pd
df=pd.DataFrame(columns=word_titles)
df
```

```python
column_data=Soup.find_all("tr")
column_data
```

```python
for row in column_data:
  row_data=row.find_all("td")
  individual_data=[data.text.strip() for data in row_data]
  print(individual_data)
```

```python
for row in column_data:
  row_data=row.find_all("td")
  individual_data=[data.text.strip() for data in row_data]
  if len(individual_data)==len(df.columns):
    df.loc[len(df)]=individual_data
```
